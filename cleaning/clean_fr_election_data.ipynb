```python
interesting_colums = ["Code du département", "Code de la commune", "Inscrits", "Abstentions", "Nuls", "Votants", \
                      "Nom", "% Voix/Exp", "Nom.1", "% Voix/Exp.1", "Nom.2","% Voix/Exp.2","Nom.3","% Voix/Exp.3", \
                      "Nom.4","% Voix/Exp.4", "Nom.5","% Voix/Exp.5","Nom.6","% Voix/Exp.6", "Nom.7","% Voix/Exp.7", \
                      "Nom.8","% Voix/Exp.8", "Nom.9","% Voix/Exp.9", "Nom.10","% Voix/Exp.10"]
data_elections = data_elections[interesting_colums]
data_elections.head()
```

## COMPUTE INSEE CODE BASE ON DEPARTEMENT AND COMMUNE CODE

```python
data_elections["Code du département"].unique()
```

Data type seems to differ inside an unique column ("Code du département"), so we format all data in string

```python
data_elections["Code du département"] = data_elections["Code du département"].map(lambda x : str(x))
data_elections["Code du département"].unique()
```

We remove COM related data (St Pierre et Miquelon, Nouvelle-Calédonie, Polynésie, Wallis&Futuna, St-Martin) + french who vote but actual live abroad. Still, we keep data from la Réunion, la Martinique, la Guadeloupe and la Guyanne

```python
masque = data_elections["Code du département"].isin(["ZZ", "ZW", "ZP", "ZM", "ZN" , "ZS", "ZX"])
data_elections = data_elections[~masque]
data_elections["Code du département"].unique()
```

Based on "Communes" and "Départements" columns, we calculate the "INSEE" code of a particular city. This will turn to be useful to merge different database in the futur. 

```python
def format_code_commune(x):
    if x < 10 :
        return "00" + str(x)
    if x < 100 :
        return "0" + str(x)
    else :
        return str(x)
def format_code_departement(x):
    if x == '2A' or x == '2B':
        return x
    if x == 'ZA' or x == 'ZB' or x == 'ZC' or x == 'ZD':
        return "97" 
    if int(x) < 10 :
        return "0" + x 
    else :
        return x 
```

```python
data_elections["Code du département"] = data_elections["Code du département"].apply(format_code_departement)
data_elections["Code de la commune"] = data_elections["Code de la commune"].apply(format_code_commune)
data_elections["Code Insee"] = data_elections["Code du département"] + data_elections["Code de la commune"]
```

## CONVERT RESULT 21,67 TO 21.65

```python
for i in range(11):
    data_elections.iloc[:,7+2*i] = data_elections.iloc[:,7+2*i].apply(lambda x : x.replace(',','.'))
```

## REVERSE THE TAB TO HAVE CANDIDATES NAMES INTO COLUMNS

```python
reverse_tab = data_elections.iloc[:,6:-1]

# MERGE ALL RESULTS IN A UNIQUE COLUMN BY USING STRING CONCATENATION
reverse_tab["tupple"] = ""
for i in range(11) :
    reverse_tab["tupple"] += reverse_tab.iloc[:,2*i] + ":" + reverse_tab.iloc[:,2*i+1]
    if i != 10:
        reverse_tab["tupple"] += ";"
```

```python
# SPLIT STRING INTO LIST 
# EACH ELEMENT OF THE LIST HAS THE FORM "CANDIDATE NAME, CANDIDATE RESULT"

reverse_tab["tupple"] = reverse_tab["tupple"].apply(lambda s: s.split(";"))
```

```python
# TRANSFORM EACH ELEMENT OF THE LIST INTO A TUPPLE (CANDIDATE NAME, CANDIDATE RESULT)
def lambda_function(l):
    def sub_lambda_function(s):
        ls = s.split(":")
        return (ls[0], float(ls[1]))
    
    return list(map(sub_lambda_function, l))

reverse_tab["tupple"] = reverse_tab["tupple"].apply(lambda_function)
```

```python
# SORT THE LIST BY CANDIDAT NAMES 
reverse_tab["tupple"] = reverse_tab["tupple"].apply(lambda s: sorted(s, key =lambda x:x[0]))
```

```python
# KEEP ONLY CANDIDATE RESULTS 
def extract_only_score(l):
    return list(map(lambda x:x[1], l))
reverse_tab["tupple"] = reverse_tab["tupple"].apply(extract_only_score)
```

```python
# CREATE A NEW DATAFRAME BASE ON THE COLUMN
candidates_list = ['LE PEN', 'MACRON','FILLON','MÉLENCHON','DUPONT-AIGNAN','HAMON','ASSELINEAU','ARTHAUD', \
                   'POUTOU','CHEMINADE','LASSALLE']
candidates_list = sorted(candidates_list)
new_df = pd.DataFrame(reverse_tab["tupple"].values.tolist(), columns=candidates_list)
```

```python
# OBTAIN THE FINAL DATAFRAME BY MERGING WITH PREVIOUS DF 
clean_fr_elections = pd.concat([data_elections[['Inscrits', 'Nuls', 'Votants', "Code Insee"]], new_df], axis=1)          
```
