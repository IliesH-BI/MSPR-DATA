```python
def ratio_na(col):
    ratio = len(col[col.isna()])/len(col)
    return ratio
def nb_na(col):
    nb = len(col[col.isna()])
    return nb
```

```python
for col in data_num:
    print('Column ', data_num[col].name, 'has ',nb_na(data_num[col]),' soit ',ratio_na(data_num[col]),'% of missing value')
```

```python
error=0
column=None
for i in range(data_num.shape[1]):
    if type(data_num[0][i])!=np.float64:
        column=i
        error+=1
print(error)
print(column)
```

```python
M=np.mean(data_num,0)
```

```python
u=data_num-M
sigma=np.dot(u.T,u)
eigval,eigvec = np.linalg.eig(sigma)
print(eigval)
```

```python
idx = eigval.argsort()[::-1]
print(idx)
```

```python
print(eigvec[0][:])
```

```python
eigvec=eigvec[:,idx]
```

```python
newData=np.dot(u,np.real(eigvec[:,:50]))
```
